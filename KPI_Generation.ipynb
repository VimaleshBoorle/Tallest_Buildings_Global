```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\adari\IdeaProjects\project\Output\ skyscapers_country\skyscapers_count.csv')
```

```python
plt.figure(figsize=(15,6))
sns.barplot(y=df['country'],x=df['number'],orient='h',palette='crest')
plt.xlabel("count")
plt.ylabel("Building Name")
plt.title("Countrywise skyscapers count")
for index, value in enumerate(df['number']):
    plt.text(value, index, str(value))
```

```python
df = pd.read_csv(r'C:\Users\adari\IdeaProjects\project\countrywise_count.csv')
country = list(df.country)
count = list(df['count'])
country_count = dict(zip(country,count))
colors = sns.color_palette('pastel')[0:5]
plt.figure(figsize=(8,6))
plt.pie(country_count.values(), labels=country_count.keys(),colors = colors, autopct='%.0f%%')
plt.title("Geographical distribution of the building")
```
